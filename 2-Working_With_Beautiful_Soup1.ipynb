<center><h1>Working with Beautiful Soup</h1></center>

```python
from bs4 import BeautifulSoup
import requests
```

```python
x = requests.get('https://vit.ac.in')
```

```python
soup = BeautifulSoup(x.text, 'html.parser')
```

#### Display Source Code

```python
print(soup.prettify())
```

#### List all Child Tags in Head and Body tag

```python
list(soup.children)
```

```python
html = soup.html
```

```python
list(html.children)
```

```python
list(html.body.children)
```

```python
body = html.body
```

```python
list(body.children)
```

#### List all a tags in the website

```python
soup.find_all('a')
```

#### Search with class Name

```python
soup.find_all(class_="nav-link")
```

```python
title_tag=soup.title
```

#### Display all reference links in the website

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
